```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import csv
%matplotlib inline
sns.set()
```

```python
# Import Datasets into Panda Dataframes
df1 = pd.read_csv('../conversion/DataAirQuality_MASTER.csv')
df2 = pd.read_csv('../conversion/DataWeather_MASTER.csv')
df1.info
#df2.info
```

```python
# Label dataset columns as their attributes
df1.columns = ['Time','PM2.5','PM10']
df1.head()
```

```python
df2.columns = ['Time','Pressure','PrecipIntensity','PrecipProbability','WindSpeed','WindGust']
df2.head()
```

```python
df1.plot()
plt.show()
```

```python
df2.plot()
plt.show()
```

```python
# Separating different data sources
PM25data = df1["PM2.5"]
PM10data = df1["PM10"]
PIdata = df2["PrecipIntensity"]
PPdata = df2["PrecipProbability"]
WSdata = df2["WindSpeed"]
WGdata = df2["WindGust"]
```

```python
# Normalising datasets and plotting
PM25data_norm = (((PM25data-PM25data.min()))/(PM25data.max()-PM25data.min()))*10
PM10data_norm = (((PM10data-PM10data.min()))/(PM10data.max()-PM10data.min()))*10
PIdata_norm = (((PIdata-PIdata.min()))/(PIdata.max()-PIdata.min()))*10
PPdata_norm = (((PPdata-PPdata.min()))/(PPdata.max()-PPdata.min()))*10
WSdata_norm = (((WSdata-WSdata.min()))/(WSdata.max()-WSdata.min()))*10
WGdata_norm = (((WGdata-WGdata.min()))/(WGdata.max()-WGdata.min()))*10
```

```python
# Plotting normalised datasets
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(PM25data_norm,'b', label = "PM2.5")
plt.plot(PM10data_norm,'g', label = "PM10")
plt.plot(PIdata_norm,'r', label = "PrecipIntensity")
plt.plot(PPdata_norm,'c', label = "PrecipProbability")
plt.plot(WSdata_norm,'y', label = "WindSpeed")
plt.plot(WGdata_norm,'m', label = "WindGust")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
# Autocorrelation for Different Data Sources
from pandas.plotting import autocorrelation_plot
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
autocorrelation_plot(PM25data)
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
autocorrelation_plot(PM10data)
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
autocorrelation_plot(PIdata)
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
autocorrelation_plot(PPdata)
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
autocorrelation_plot(WSdata)
plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
autocorrelation_plot(WGdata)
```

```python
# Correlation Coefficients between Data Sources
C_PM10_PI = PM10data.corr(PIdata)
print("C_PM10_PI: ", C_PM10_PI)
C_PM10_PP = PM10data.corr(PPdata)
print("C_PM10_PP: ", C_PM10_PP)
C_PM10_WS = PM10data.corr(WSdata)
print("C_PM10_WS: ", C_PM10_WS)
C_PM10_WG = PM10data.corr(WGdata)
print("C_PM10_WG: ", C_PM10_WG)
C_PM25_PI = PM10data.corr(PIdata)
print("C_PM25_PI: ", C_PM25_PI)
C_PM25_PP = PM10data.corr(PPdata)
print("C_PM25_PP: ", C_PM25_PP)
C_PM25_WS = PM10data.corr(WSdata)
print("C_PM25_WS: ", C_PM25_WS)
C_PM25_WG = PM10data.corr(WGdata)
print("C_PM25_WG: ", C_PM25_WG)
```
